## Getting the data

```python
# create dataframe of coin prices for a year

from pandas_datareader import data as pdr
import pandas as pd

dates = ["2021-01-23", "2022-01-22"]

coins = ["DOGE-USD", "BTC-USD", "SHIB-USD", "XTZ-USD", "XLM-USD", "COVAL-USD", "FET-USD"]
dg_coin = {}

coin_dict = {}
for i, c in enumerate(coins):
    data = pdr.get_data_yahoo(c, start=dates[0], end=dates[1])
    prices = data['High'] / 2 + data['Low'] / 2
    coin_dict[c] = prices
    
df = pd.DataFrame(coin_dict)
```

```python
df
```

## Simple Year Visualization of Crypto

```python
import matplotlib.pyplot as plt
```

```python
for i in df:
    plt.plot(df.index, df[i])
    plt.xlabel(i)
    plt.show()
```

## Finding What Day of the Week Works Best

```python
df['WeekDay'] = [i.weekday() for i in df.index]
df
```

```python
plt.plot(df.index[df['WeekDay'] == 0], df['DOGE-USD'][df['WeekDay'] == 0])
```
